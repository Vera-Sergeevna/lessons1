Задание 1.
Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером. Если да, то функция должна возвращать отдельно номер и регион.

```python
import re
```

```python

def number(user_number):
    number_re = r'[АВЕКМНОРСТУХABEKMHORCTYX]{1}[0-9]{3}[АВЕКМНОРСТУХABEKMHORCTYX]{2}[0-9]{2,3}'
    number_try = re.fullmatch(number_re, user_number, flags=re.I)
    if number_try != None:
        return number_try.group()[0:6],number_try.group()[6:]
    else:
        print('Неверный формат')
    
number('X007AR35')
```

Задание 2.
Постарайтесь написать (в крайнем случае найти) самый полный шаблон для проверки e-mail адресов через регулярное выражение, которая проверяет e-mail адрес на валидность. Можно руководствоваться следующими минимальными требованиями:
• адрес должен содержать специальный символ “@”, который отделяет имя пользователя почтовой системы от доменного имени;
• адрес не должен содержать следующие символы: “пробел”, “,”, “:”, “;”, “!”, “#”, “%”, “*”, “(”, “)”, “=”, “+”, “{”, “}”, " " ", " ’ ", " / ", " \ " и “|”;
• адрес должен состоять только из латинских символов;
• после символа “@” должна быть как минимум одна “.”;
• после последней точки должно быть не менее 2-х и не более 4-х символов, причем наличие цифр не допускается;
• между последней точкой и символом “@” должно быть не менее 2-х символов
• слева от “@” должно быть не менее четырех символов

```python
import re

list_mail = ['push.1980k@mail.ru',
             'irochka56@yandex.ru',
             'pops@aba@jmail.com' ,
             'noName@856.78' 
]

def mail_check(inp):
    mail_name = r'[a-z\d\.\-]{4,}\@[a-z]{2,}\.[a-z]{2,4}'
    mail_proverka = re.fullmatch(mail_name, inp, flags=re.I)
    if mail_proverka != None:
        print(mail_proverka.group())

for mails in list_mail:
    mail_check(mails)
```

**Задание 3. **
Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений.

```python
import re


string = "Ты Ты  родедендроны родедендроны полил полил "

print(re.sub(r'\b([^\W\d_]+)(\s+\1)+\b', r'\1', re.sub(r'\W+', ' ', string).strip(), flags=re.I))
```

Задание 4.
Напишите функцию, которая будет проверять номер сотового телефона на валидность, если он валиден, то переводить его в формат:
+7(xxx)-xxx-xx-xx
​
Постарайтесь предусмотреть как можно больше адекватных форматов изначального ввода номера. Примеры валидных форматов:
​
• +7 955 555-55-55;
• 9555555555;
• 8(955)555-55-55;
• +7 955 555 55 55;
• 7(955) 555-55-55;
• +7 955+555+55+55.

```python
import re
tel_nums = [
    '+7 900 100-33-33', '9555555555', '(955)552-55-55', '+7 955 555 55 55',
    '7(955) 555-55-55', '7 955+555+55+55', '+79261234567', '89261234567',
    '79261234567', '+7 926 123 45 67', '8(926)123-45-67', '732123-45-67',
    '9261234567', '79261234567', '(495)1234567', '(495) 123 45 67'
]
def tel_check(phone):
    num_pattern = r'((8|\+?7)[\- ]?)?(\(?\d{3}\)?[\+\- ]?)?[\d\+\- ]{7,10}'
    number = re.fullmatch(num_pattern, phone)
    print(f'Вы ввели номер - {phone}')
    if number != None:
        telef = re.sub(r'[- _():=+]', r'', number.group())
        return re.sub(r'\d?(\d\d\d)(\d\d\d)(\d\d)(\d\d)', r'+7(\1)-\2-\3-\4',
                      telef)
    else:
        return 'Ошибка'
for numb in tel_nums:
    print(tel_check(numb))
```
